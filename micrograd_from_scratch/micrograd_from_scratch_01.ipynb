```python
import math
import numpy
import matplotlib

%matplotlib inline
numpy.set_printoptions(suppress=True)
```

```python
# Variables
point = 3.0
```

```python
# f(x) = 3x^2 - 4x + 5
def f(x):
    return 3 * pow(x, 2) - 4 * x + 5


def f_prime(x):
    return 6 * x - 4
```

### Step-by-Step Derivative

The **derivative** measures how much f(x) changes when x changes by a tiny amount.

**Sample function:** $f(x) = 3x^2 - 4x + 5$

**Rules used:**

- **Power rule:** the derivative of $x^n$ is $n \cdot x^{n-1}$
- **Constant multiple rule:** the derivative of $c \cdot g(x)$ is $c \cdot g'(x)$
- **Sum rule:** differentiate each term separately
- **Constant rule:** the derivative of a constant is $0$

**Apply term by term:**

| Term   | Rule applied                                | Derivative              |
| ------ | ------------------------------------------- | ----------------------- |
| $3x^2$ | Power rule: bring down 2, reduce power by 1 | $3 \cdot 2x^{2-1} = 6x$ |
| $-4x$  | Power rule ($x = x^1$): bring down 1        | $-4 \cdot 1 = -4$       |
| $5$    | Constant rule                               | $0$                     |

**Result:** $f'(x) = 6x - 4$

**Quick verification:**

At $x = 3$: $f'(3) = 6(3) - 4 = 14$

We can check numerically: for a tiny $h$, $\frac{f(x+h) - f(x)}{h}$ should be close to $14$.

```python
xs = numpy.arange(-5, 5, 0.25)
slope = f_prime(point)
y_at_point = f(point)

# Tangent line: y = f(point) + f'(point) * (x - point)
tangent_xs = numpy.linspace(point - 2, point + 2, 50)
tangent_ys = y_at_point + slope * (tangent_xs - point)

matplotlib.pyplot.plot(xs, f(xs), label="f(x)")
matplotlib.pyplot.plot(tangent_xs, tangent_ys, "--", label=f"f'(x)")
matplotlib.pyplot.plot(point, y_at_point, "ro", markersize=4, label=f"point")
matplotlib.pyplot.legend()
matplotlib.pyplot.title(f"Function: f({point}) = {y_at_point}. Slope: f'({point}) = {slope}")
```

```python
h = 0.00000001
x = 3.0
print(f"{f(x):.30f}")
print(f"{f(x + h):.30f}")
print(f"{f(x + h) -f(x):.30f}")
print(f"{(f(x + h) -f(x)) / h:.30f}")
print(f"{f_prime(x):.30f}")
```

```python
h = 0.0001

a = 2.0
b = -3.0
c = 10.0

d1 = a * b + c

a += h
d2 = a * b + c

print("d1: ", d1)
print("d2: ", d2)
print("slope: ", (d2 - d1) / h)
```
